```python
def write_images(soup):
    global img_id
    a=[]
    for img in soup.find_all("img"):
        image=img.get("src")
        
        if "svg" not in image and image not in a:
            a.append(image)
        
            imagefile=open(str(img_id)+".jpg",'wb')
            imagefile.write(urllib.request.urlopen(image).read())
            print(imagefile)
            imagefile.close()
            img_id+=1
    
```
